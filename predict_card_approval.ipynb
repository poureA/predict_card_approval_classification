```python
with open('F:\\datasets\\cc_approvals.data','r') as file :
    val = file.readlines()
base = []
for i in val :
    x = i.split(',')
    base.append(x)
```

```python
base[0:11]
```

```python
import pandas as pd 
pd_file = pd.DataFrame(base)
pd_file.sample(10)
```

```python
pd_file.info()
```

```python
label = []
for i in pd_file[15] :
    label.append(i[0])
pd_file = pd_file.drop([15],axis=1)
pd_file['label'] = label
```

```python
pd_file.sample(10)
```

```python
for i in pd_file.columns :
    print(i)
    print(pd_file[i].value_counts(ascending=False))
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer_q = SimpleImputer(missing_values='?',strategy='constant',fill_value=np.nan)
imputed_df_q = pd.DataFrame(imputer_q.fit_transform(pd_file),columns=pd_file.columns)
imputed_df_q.info()
```

```python
imputed_df_q.sample(10)
```

```python
lbl = imputed_df_q['label']
imputed_df_q = imputed_df_q.drop(['label'],axis=1)
imputed_df_num = imputed_df_q.drop([0,3,4,5,6,8,9,11,12],axis=1)
imputed_df_num.head()
```

```python
base = []
child = []
for i in imputed_df_num.columns :
    for j in imputed_df_num[i] :
        if j == np.nan :
            val = [np.nan]
            child.append(val)
            continue
        x = float(j)
        val = [x]
        child.append(val)
    base.append(child)
    child = []
df_list = []
col = imputed_df_num.columns
c = 0
for i in base :
    df = pd.DataFrame(i,columns=[col[c]])
    df_list.append(df)
    c +=1
floated_imputed_df_num = pd.concat([i for i in df_list],axis=1)
floated_imputed_df_num
```

```python
floated_imputed_df_num.dtypes
```

```python
imputer_mean = SimpleImputer(missing_values=np.nan,strategy='mean')
imputed_ready = pd.DataFrame(imputer_mean.fit_transform((floated_imputed_df_num)),columns=floated_imputed_df_num.columns)
imputed_ready.isnull().sum()
```

```python
imputed_ready.info() #read to concat
```

```python
imputed_df_nan = imputed_df_q.drop(imputed_df_num.columns,axis=1)
imputed_df_nan.head()
```

```python
imputed_df_nan.isnull().sum()
```

```python
for i in imputed_df_nan.columns :
    print(i)
    print(imputed_df_nan[i].value_counts(ascending=False))
```

```python
imputer_mf = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputed_df_nan = pd.DataFrame(imputer_mf.fit_transform(imputed_df_nan),columns=imputed_df_nan.columns)
imputed_df_nan.info()
```

```python
encoded_nan = pd.get_dummies(imputed_df_nan)
encoded_nan.describe()
```

```python
encoded_nan.info()
```

```python
finall_df = pd.concat([imputed_ready,encoded_nan],axis=1)
finall_df.info()
```

```python
from sklearn.preprocessing import StandardScaler , LabelEncoder
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(finall_df),columns=finall_df.columns)
scaled_df.sample(10)
```

```python
encoder = LabelEncoder()
encoded_lbl = encoder.fit_transform(lbl)
```

```python
from sklearn.model_selection import train_test_split
xtr , xts , ytr , yts = train_test_split(scaled_df,encoded_lbl,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
alg = SVC().fit(xtr,ytr)
alg.score(xtr,ytr)*100
```

```python
alg.score(xts,yts)*100
```

```python
alg.score(scaled_df,encoded_lbl)*100
```

```python
scores = []
for i in range(10):
    xtr , xts , ytr , yts = train_test_split(scaled_df,encoded_lbl,test_size=0.2,shuffle=True)
    alg = SVC().fit(xtr,ytr)
    train = alg.score(xtr,ytr)*100
    scores.append(train)
    test = alg.score(xts,yts)*100
    scores.append(test)
    total = alg.score(scaled_df,encoded_lbl)*100
    scores.append(total)
np.mean(scores)
```
